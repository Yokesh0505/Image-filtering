### Introduce a Gaussian noise and speckle noise over image

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
from skimage.util import random_noise
```

```python
fig=plt.figure(figsize=(30,30))

I = cv2.imread('wp.jpg', 1); # 1/ -1: color mode; 0: gray mode
gauss = random_noise(I, mode='gaussian', seed=None, clip=True)
sp = random_noise(I, mode='speckle', seed=None, clip=True)

fig.add_subplot(1,3,1), plt.imshow(I), plt.title('Origin'),plt.axis('off')
fig.add_subplot(1,3,2), plt.imshow(gauss), plt.title('Gaussian'),plt.axis('off')
fig.add_subplot(1,3,3), plt.imshow(sp), plt.title('Speckle'),plt.axis('off')
plt.show();
```

### Perform image enhancement using spatial domain filters

The spatial domain is used to define the actual spatial coordinates of pixels within an image, so when we use this term in the image enhancement business, were talking about things like equalization, smoothing, and sharpening.`m

```python
# read a image using imread
img = cv2.imread('wp.jpg', 0)

# creating a Histograms Equalization
# of a image using cv2.equalizeHist()
equ = cv2.equalizeHist(img)

# stacking images side-by-side
res = np.vstack(equ)

# show image input vs output
cv2.imshow('wp.jpg',res)

plt.imshow(res)
```

### Program to comparing averaging & median filter

The Averaging Filter technique takes the average of all the pixels under the kernel area and replaces the central element. The functions cv2.blur() and cv2.boxFilter() can be used to perform the averaging filter.

```python
#cv2.GaussianBlur(src, ksize, sigma_x, dst, sigma_y, border_type)
```

```python
path = r'wp.jpg'

# using imread()  
img = cv2.imread(path)

dst = cv2.GaussianBlur(img,(5,5),cv2.BORDER_DEFAULT) 

cv2.imshow('image', np.hstack((img, dst)))

plt.imshow(dst)
```

```python
#cv2.boxFilter(src, dst, depth, ksize, anchor, normalize, bordertype)
```

```python
path = 'wp.jpg'

# using imread()  
img = cv2.imread(path)

im1 = cv2.blur(img,(1,1))
im2 = cv2.boxFilter(img, -1, (2, 2), normalize=True)  

cv2.imshow('image', np.hstack((im1, im2)))
cv2.waitKey(0);
cv2.destroyAllWindows();
cv2.waitKey(1)
```

Python OpenCV provides the cv2.medianBlur() function to blur the image with a median kernel. 
This is a non-linear filtering technique. It is highly effective in removing salt-and-pepper noise. 
This takes the median of all the pixels under the kernel area and replaces the central component with this median value.

```python
#cv2.medianBlur(image, ksize)
```

```python
path = r'wp.jpg'

# using imread()  
img = cv2.imread(path)
dst = cv2.medianBlur(img,7)

cv2.imshow('image', np.hstack((img, dst)))
cv2.waitKey(0);
cv2.destroyAllWindows();
cv2.waitKey(1)
```

### Sharpen an image using high pass filter in frequency domain

```python
def main():
    
    imgpath = r"wp.jpg"
    img = cv2.imread(imgpath, 1)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    edges = cv2.Laplacian(img, -1, ksize=5, scale=1, delta=0, 
                          borderType=cv2.BORDER_DEFAULT)

    output = [img, edges]
    titles = ['Original', 'Edges']
    
    for i in range(2):
        plt.subplot(1, 2, i+1)
        plt.imshow(output[i], cmap = 'gray')
        plt.title(titles[i])
        plt.xticks([])
        plt.yticks([])
    plt.show()

if __name__ == "__main__":
    main()
```

```python
def main():
    
    imgpath = "wp.jpg"
    img = cv2.imread(imgpath, 1)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    edgesx = cv2.Sobel(img, -1, dx=1, dy=0, ksize=11, scale=1,
                       delta=0, borderType=cv2.BORDER_DEFAULT)
    
    edgesy = cv2.Sobel(img, -1, dx=0, dy=1, ksize=11, scale=1,
                       delta=0, borderType=cv2.BORDER_DEFAULT)
    
    edges = edgesx + edgesy

    output = [img, edgesx, edgesy, edges]
    titles = ['Original', 'dx=1 dy=0', 'dx=0 dy=1', 'Edges']
    
    for i in range(4):
        plt.subplot(2, 2, i+1)
        plt.imshow(output[i], cmap = 'gray')
        plt.title(titles[i])
        plt.xticks([])
        plt.yticks([])
    plt.show()

if __name__ == "__main__":
    main()
```
